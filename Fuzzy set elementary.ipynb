# Fuzzy sets

## Basic Set properties

```
import numpy as np
```

```
f_set=dict(zip((1,2,3,4,5,6,7),0.1*np.array([10,2,3,4,5,6,7])))
```

```
f_set
```

```
def height(f_set):
    height1={}
    for i in f_set.keys():
        if f_set[i]==1.0:
            height1[i]=f_set[i]
    return height1
```

```
height(f_set)
```

```
def core(f_set):
    return(height(f_set).keys())
```

```
core(f_set)
```

```
cut=0.5
def crossover(f_set,cut):
    cross_over={}
    for i in f_set.keys():
        if f_set[i]==cut:
            cross_over[i]=f_set[i]
    return cross_over
```

```
def bandwidth(f_set):
    cross=crossover(f_set,0.5)
    band=np.max(np.array(cross.keys()))-np.min(np.array(cross.keys()))
    return band
```

```
def alpha_cut(f_set,cut):
    strong_alpha={}
    weak_alpha={}
    for i in f_set.keys():
        if f_set[i]>cut:
            strong_alpha[i]=f_set[i]
        else:
            weak_alpha[i]=f_set[i]
    return strong_alpha,weak_alpha
```
