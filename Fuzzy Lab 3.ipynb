##### Min-max Composition


```python
def minmax(A,B):
    maximum=[]
    for i in range(len(A)):
        maximum.append(max(A[i],B[i]))
    return min(maximum)
```

```python
A=[1,2,4,5,6]
B=[4,1,5,7,8]
minmax(A,B)
```

##### Max-product Composition


```python
def maxprod(r1,r2):
    product=[]
    for i in range(len(r1)):
        product.append(r1[i]*r2[i])
    return max(product)
```

```python
r1=[0.4,0.2,0.8,0.9]
r2=[0.9,0.2,0.5,0.7]
maxprod(r1,r2)
```
