##### Max-min Composition


```python
def maxmin(r1,r2):
    mini=[]
    for i in range(len(r1)):
        mini.append(min(r1[i],r2[i]))
    return max(mini)

r1=[0.6,0.8,0.3,0.2]
r2=[0.9,0.2,0.5,0.7]

maxmin(r1,r2)
```

##### Max-product Composition


```python
def maxprod(r1,r2):
    mul=[]
    for i in range(len(r1)):
        mul.append(r1[i]*r2[i])
    return max(mul)

r1=[0.6,0.8,0.3,0.2]
r2=[0.9,0.2,0.5,0.7]

maxprod(r1,r2)
```
