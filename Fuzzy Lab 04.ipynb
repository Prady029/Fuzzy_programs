```python
temp=[[16,0.4],[18,0.8],[20,1.0],[22,1.0],[24,0.8],[26,0.5]]
humi=[[0,0.2],[20,0.8],[40,1],[60,0.6],[80,0.2]]
```

```python
def Fuzzy_union(a,b):
    unn=a+b
    unn.sort()
    uni=[]
    for i in range(1,len(unn)):
        if(unn[i][0]!=unn[i-1][0]):
            uni.append(unn[i-1])
    for i in range(1,len(unn)):
        if(unn[i][0]==unn[i-1][0] and unn[i][1]>unn[i-1][1] and unn[i] not in uni):
            uni.append(unn[i])
    uni.sort()    
    return uni
```

```python
#lets assume that acceptable membership value is atleast c
c=0.5
acctemp=[]
acchumi=[]

for i in range(len(temp)):
    if(temp[i][1]>=c):
        acctemp.append(temp[i])
        
for i in range(len(humi)):
    if(humi[i][1]>=c):
        acchumi.append(humi[i])

accTempOrAccHumi = Fuzzy_union(acctemp,acchumi)
```
